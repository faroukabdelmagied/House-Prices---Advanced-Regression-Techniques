## Importing Libraries

```
## Importing the necessary libraries:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

## Display all coloumns of the data frame:
pd.pandas.set_option('display.max_columns', None)
```

```
## Reading the data set
dataset = pd.read_csv('train.csv')

## Printing the dataset shape:
print(dataset.shape)
```

```
## Print the top 5 records:
dataset.head()
```

## Missing Values

```
## Make a list of features that have missing values:
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]
features_with_na

## Print the feature name with the percentage of missing values:
for feature in features_with_na:
  print(feature, np.round(dataset[feature].isnull().mean(),4), '% missing values')
```

### Finding the Relationship of Missing values with SalesPrice

```
## Create a copy of the data set:
for feature in features_with_na:
  data = dataset.copy()

  ## Lets make a variable that indicates 1 if the observation was missing or zero if the observations was not missing
  data[feature] = np.where(data[feature].isnull(), 1, 0)

## Setting the colors for the bar plot:
  c = ['blue', 'red']
  ## Now let us calculate the mean SalePrice where the information is missing or present:
  data.groupby(feature)['SalePrice'].median().plot.bar(color = c)
  plt.title(feature)
  plt.ylabel('SalePrice (median)')
  plt.xlabel('')
  plt.show()
```

## Numerical Features

```
## Extracting the numerical features from the data:
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']

print('Number of numerical variables:', len(numerical_features))

## Printing the head of our data frame:
dataset[numerical_features].head()
```

### Temporial Features

```
year_feature= [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_feature
```

```
## Let us explore the content of these year variables
for feature in year_feature:
  print(feature, dataset[feature].unique())
```

####  Temporal Datetime Variables Analysis

```
## First we will check if there is a relationship between the year the house was sold and the target variable (YrSold):
dataset.groupby('YrSold')['SalePrice'].median().plot(figsize = (15,10), color = 'blue')
plt.xlabel('Year Sold', size=13)
plt.ylabel('Sale Price (median)', size=13)
plt.title('House Price vs YearSold', size=15)
```

```
## Here we will compare the difference between all the year features and the SalePrice:

for feature in year_feature:
  if feature != 'YrSold':
    data = dataset.copy()

    ## We will capture the difference between the year variable and the year sold
    data[feature] = data['YrSold'] - data[feature]

    plt.scatter(data[feature],data['SalePrice'], color = 'blue')
    plt.xlabel(feature, size = 13)
    plt.ylabel('SalePrice', size = 13)
    plt.show()
```

### Discrete Features

```
discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print('Discrete Variables Count:', len(discrete_feature))
```

```
discrete_feature
```

```
dataset[discrete_feature].head()
```

```
## Plotting the discrete variable:
for feature in discrete_feature:
  data = dataset.copy()
  data.groupby(feature)['SalePrice'].median().plot.bar(color='blue')
  plt.xlabel(feature, size=13)
  plt.ylabel('SalesPrice', size=13)
  plt.title('')
  plt.show()
```

### Continuous Features

```
continuous_feature = [feature for feature in numerical_features if feature not in discrete_feature+year_feature+['Id']]

## Print the length of continuous variables:
print('Continuous features:', len(continuous_feature))
```

```
## Plotting the continuous features:
for feature in continuous_feature:
  data = dataset.copy()
  data[feature].hist(bins=25,color = 'blue')
  plt.xlabel(feature, size=13)
  plt.ylabel("Count", size=13)
  plt.title('')
  plt.show()
```

#### Converting Skewed Data

```
# We will be using logarithmic transformation:
for feature in continuous_feature:
  data = dataset.copy()
  if 0 in data[feature].unique():
    pass
  else:
    data[feature]=np.log(data[feature])
    data['SalePrice ']=np.log(data['SalePrice'])
    plt.scatter(data[feature], data['SalePrice'], color = 'blue')
    plt.xlabel(feature, size=13)
    plt.ylabel('SalePrice', size=13)
    plt.title('')
    plt.show()
```

#### Dealing with Outliers

```
for feature in continuous_feature:
  data=dataset.copy()
  if 0 in data[feature].unique():
    pass
  else:
    data[feature]=np.log(data[feature])
    data.boxplot(column=feature)
    plt.ylabel(feature)
    plt.title(feature)
    plt.show()
```
