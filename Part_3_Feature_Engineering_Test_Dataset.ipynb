# Feature Engineering - Test Dataset

#### Steps to perform Feature Engineering:
  1. Missing Values
  2. Temporal Variables
  3. Categorical Variables
  4. Standarise the values of the variables to the same range

## Preprocessing and Importing Libraries

```
## Importing the needed libraries:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
## To visualise all the columns in the dataframe
pd.pandas.set_option('display.max_columns', None)
```

```
## Let us read the test data set first:
dataset_test = pd.read_csv('test.csv')
dataset_test.head()
```

```
dataset_test.info()
```

## 2. Missing Data

### Categorical Features

```
## Taking care of missing values in categorical features: fill NA with the "missing category instead"

## Select categorical features with missing values:
categorical_feat_nan = [feature for feature in dataset_test.columns if dataset_test[feature].isnull().sum()>=1 and dataset_test[feature].dtype=='O']
categorical_feat_nan

## Calculate the % of missing values in these features:
for feature in categorical_feat_nan:
  print('{} :{}% missing values'.format(feature, np.round(dataset_test[feature].isnull().mean()*100,2)))
```

```
## Replacing categorical features:
def replace_catog_feature(dataset_test, categorical_feat_nan):
  data=dataset_test.copy()
  data[categorical_feat_nan]=data[categorical_feat_nan].fillna('Missing')

  return data
dataset_test=replace_catog_feature(dataset_test, categorical_feat_nan)
```

```
## Checking if there is missing categorical features
dataset_test[categorical_feat_nan].isnull().sum()
```

```
## Converting MSubclass from numbers into Categories
dataset_test['MSSubClass'].replace({20:'A',
                          30:'B',
                          40:'C',
                          45:'D',
                          50:'E',
                          60:'F',
                          70:'G',
                          75:'H',
                          80:'I',
                          85:'J',
                          90:'K',
                          120:'L',
                          150:'M',
                          160:'N',
                          180:'O',
                          190:'P'}, inplace=True)
```

```
dataset_test.head()
```

## Numerical Features

```
## Now let us check numerical variables that have missing values:
numerical_test_nan = [feature for feature in dataset_test.columns if dataset_test[feature].isnull().sum()>=1 and dataset_test[feature].dtype!='O']

## Presenting the amount of missing values:
for feature in numerical_test_nan:
  print('{} :{}%'.format(feature, np.round(dataset_test[feature].isnull().mean()*100,2)))
```

```
## Filling the missing numerical data with the mean value:
for feature in numerical_test_nan:
  median_value = dataset_test[feature].median()
  dataset_test[feature].fillna(median_value, inplace=True)

## Checking for numerical missing values:
dataset_test[numerical_test_nan].isnull().sum()
```

## Temporal Variables (Date Time Variables)

```
## Temporal Variables
dataset_test[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

```
## Converting temporal variables to the number of years:
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
  dataset_test[feature]=dataset_test['YrSold']-dataset_test[feature]
```

```
## Exploring the conversion:
dataset_test[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

## Transforming Skewed Numerical Variables (Log)
Since the numerical variables are skewed we will perform log normal distribution.

```
## Numerical features that are skewed and that have no zeros within them:
num_features_test=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea']
```

```
## Apply the log transformation:
for feature in num_features_test:
  dataset_test[feature]=np.log(dataset_test[feature])
```
